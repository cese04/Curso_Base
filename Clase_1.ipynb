## Python




Python es hoy en día uno de los lenguajes que mejor han sido recividos para el desarrollo de técnicas de Machine Learning gracias a varias características, entre estas tenemos:

* Facilidad de interpretación: La filosofía detrás de Python es la de la simplicidad, la función y tarea de un código ha de entenderse al leerlo.
* Lenguaje de propósito general: Python puede ser usado para diversas tareas como desarrollo web, creación de aplicaciones y cómputo científico.
* Multiplataforma: Los códigos de Python pueden ser interpretados por Linux, Windows y OSX.

### ¿Por qué usar Python?

* No queremos reprogramar cómo plotear una curva o hacer una transformada de Fourier.
* Facilidad: las ciencias computacionales no son nuestro enfoque o nuestra preparación.
* El código debe ser fácil de comprender
* Un código eficiente se ejecuta rápido. Pero esto se vuelve inútil si requerimos dedicar mucho tiempo a la programación del mismo.
* Un solo lenguaje para todo.
  * Django: Desarrollo Web.
  * Beautiful Soup y Requests: Usados para REST API y data mining.
  * GPIO: Usada en computadoras de placa reducidad (Raspeberry Pi, Orange Pi, etc.) para controlar  puertos y PWM.
  * PySerial: Comunicación de dispositivos por puerto serial (Arduino, PIC, Raspberry).
  * PyCuda: Programación sobre tarjetas GPU Nvidia.
